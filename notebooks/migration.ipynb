```python
import pandas as pd
import os

# ==========================================================
# Create Reports Folder
# ==========================================================
os.makedirs("reports", exist_ok=True)

# ==========================================================
# Load Risk Report
# ==========================================================
df = pd.read_csv("reports/risk_report.csv")

print("="*75)
print("          POST-QUANTUM CRYPTOGRAPHY MIGRATION STRATEGY")
print("="*75)

# ==========================================================
# Migration Priority
# ==========================================================

def migration_priority(risk, years):

    if risk == "High" and years >= 10:
        return "Priority 1"

    elif risk == "High":
        return "Priority 2"

    elif risk == "Low":
        return "Priority 3"

    return "Priority 4"

# ==========================================================
# Migration Phase
# ==========================================================

def migration_phase(priority):

    phases = {
        "Priority 1": "Phase 1 - Immediate Migration",
        "Priority 2": "Phase 2 - Hybrid Deployment",
        "Priority 3": "Phase 3 - Planned Migration",
        "Priority 4": "Phase 4 - Monitor & Review"
    }

    return phases.get(priority, "Manual Review")

# ==========================================================
# Hybrid Deployment
# ==========================================================

def hybrid_strategy(algorithm):

    if algorithm == "RSA":
        return "RSA + ML-KEM"

    elif algorithm == "ECC":
        return "ECC + ML-DSA"

    elif algorithm == "AES":
        return "AES-256"

    return "Manual Review"

# ==========================================================
# NIST Alignment
# ==========================================================

def nist_alignment(algorithm):

    if algorithm == "RSA":
        return "Transition to ML-KEM"

    elif algorithm == "ECC":
        return "Transition to ML-DSA"

    elif algorithm == "AES":
        return "AES-256 Approved"

    return "Review Required"

# ==========================================================
# Timeline
# ==========================================================

def timeline(priority):

    schedule = {
        "Priority 1":"0-6 Months",
        "Priority 2":"6-12 Months",
        "Priority 3":"12-24 Months",
        "Priority 4":"Beyond 24 Months"
    }

    return schedule.get(priority,"TBD")

# ==========================================================
# KPI
# ==========================================================

def success_metric(priority):

    metrics = {
        "Priority 1":"100% Migration",
        "Priority 2":"Hybrid Deployment Active",
        "Priority 3":"Pilot Completed",
        "Priority 4":"Annual Review"
    }

    return metrics.get(priority,"Review")

# ==========================================================
# Lists
# ==========================================================

priority_list = []
phase_list = []
hybrid_list = []
nist_list = []
timeline_list = []
kpi_list = []

# ==========================================================
# Process Assets
# ==========================================================

for _, row in df.iterrows():

    priority = migration_priority(
        row["RiskLevel"],
        row["DataLifeYears"]
    )

    priority_list.append(priority)

    phase_list.append(
        migration_phase(priority)
    )

    hybrid_list.append(
        hybrid_strategy(row["Algorithm"])
    )

    nist_list.append(
        nist_alignment(row["Algorithm"])
    )

    timeline_list.append(
        timeline(priority)
    )

    kpi_list.append(
        success_metric(priority)
    )

# ==========================================================
# Add Columns
# ==========================================================

df["MigrationPriority"] = priority_list
df["MigrationPhase"] = phase_list
df["HybridDeployment"] = hybrid_list
df["NISTAlignment"] = nist_list
df["Timeline"] = timeline_list
df["SuccessMetric"] = kpi_list

# ==========================================================
# Save Report
# ==========================================================

df.to_csv(
    "reports/migration_strategy.csv",
    index=False
)

# ==========================================================
# Console Summary
# ==========================================================

print("\n")
print("="*75)
print("PHASED MIGRATION ROADMAP")
print("="*75)

print("""
PHASE 1 (0-6 Months)
--------------------
• Identify cryptographic inventory
• Migrate Priority 1 assets
• Deploy Hybrid TLS
• Protect long-term confidential data

PHASE 2 (6-12 Months)
---------------------
• Deploy ML-KEM Hybrid Key Exchange
• Deploy ML-DSA Hybrid Signatures
• Upgrade PKI Infrastructure

PHASE 3 (12-24 Months)
-----------------------
• Complete migration of remaining systems
• Retire legacy RSA/ECC
• Validate interoperability

PHASE 4 (Beyond 24 Months)
--------------------------
• Continuous monitoring
• Annual cryptographic review
• Compliance verification
""")

print("="*75)
print("NIST ALIGNMENT")
print("="*75)

print("""
✔ Cryptographic Inventory Assessment

✔ Algorithm Agility

✔ Hybrid Cryptographic Deployment

✔ Secure Key Management

✔ Continuous Monitoring

✔ Governance and Compliance
""")

print("="*75)
print("KEY PERFORMANCE INDICATORS (KPIs)")
print("="*75)

print("""
• 100% High-Risk Assets Migrated

• 100% Hybrid TLS Deployment

• Zero Unsupported RSA Systems

• Annual Cryptographic Review Completed

• HSM Coverage Above 90%

• Zero Critical Key Management Incidents
""")

print("="*75)
print("Migration Strategy Report Generated Successfully")
print("="*75)

print("\nSaved File:")
print("reports/migration_strategy.csv")
```
