```python
import pandas as pd
import os

# -----------------------------------
# Create Reports Folder
# -----------------------------------
os.makedirs("reports", exist_ok=True)

# -----------------------------------
# Load Risk Report
# -----------------------------------
df = pd.read_csv("reports/risk_report.csv")

print("=" * 70)
print("        POST-QUANTUM CRYPTOGRAPHIC ARCHITECTURE")
print("=" * 70)


# -----------------------------------
# Recommend Hybrid Architecture
# -----------------------------------
def recommend_pqc(algorithm):

    algorithm = algorithm.upper()

    if algorithm == "RSA":
        return "Hybrid TLS (RSA + ML-KEM)"

    elif algorithm == "ECC":
        return "Hybrid Signature (ECC + ML-DSA)"

    elif algorithm == "AES":
        return "Continue AES-256"

    else:
        return "Manual Review Required"


# -----------------------------------
# Build Architecture Table
# -----------------------------------
future_architecture = []

for index, row in df.iterrows():

    recommendation = recommend_pqc(row["Algorithm"])

    future_architecture.append(recommendation)

    print("\n----------------------------------------")
    print("Asset               :", row["Asset"])
    print("Current Algorithm   :", row["Algorithm"])
    print("Future Architecture :", recommendation)


# -----------------------------------
# Add Column
# -----------------------------------
df["FutureArchitecture"] = future_architecture

# -----------------------------------
# Save Architecture Report
# -----------------------------------
df.to_csv("reports/pqc_architecture.csv", index=False)

print("\n")
print("=" * 70)
print("Architecture Design Completed Successfully")
print("Saved: reports/pqc_architecture.csv")
print("=" * 70)
```
