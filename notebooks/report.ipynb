```python
import pandas as pd
import os
from datetime import datetime
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Table,
    TableStyle
)
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet

# -----------------------------
# Create reports folder
# -----------------------------
os.makedirs("reports", exist_ok=True)

# -----------------------------
# Load Risk Report
# -----------------------------
df = pd.read_csv("reports/risk_report.csv")

# -----------------------------
# Statistics
# -----------------------------
total_assets = len(df)
high_risk = len(df[df["RiskLevel"] == "High"])
low_risk = len(df[df["RiskLevel"] == "Low"])
avg_score = round(df["RiskScore"].mean(), 2)

# Harvest Now Decrypt Later
hndl = len(df[df["HarvestNowDecryptLater"] == "YES"])

# -----------------------------
# Create PDF
# -----------------------------
pdf = SimpleDocTemplate("reports/QuantumSafe_Report.pdf")

styles = getSampleStyleSheet()

elements = []

# -----------------------------
# Title
# -----------------------------
elements.append(
    Paragraph(
        "<b><font size=20>QuantumSafe Analyzer</font></b>",
        styles["Title"]
    )
)

elements.append(
    Paragraph(
        "<b>Cryptographic Risk Assessment Report</b>",
        styles["Heading2"]
    )
)

elements.append(
    Paragraph(
        f"Generated on: {datetime.now().strftime('%d %B %Y %H:%M')}",
        styles["Normal"]
    )
)

elements.append(Spacer(1,20))

# -----------------------------
# Executive Summary
# -----------------------------
elements.append(
    Paragraph(
        "<b>Executive Summary</b>",
        styles["Heading2"]
    )
)

summary = f"""
This assessment analyzed the organization's cryptographic inventory to
identify systems vulnerable to quantum computing attacks.

<b>Total Assets:</b> {total_assets}<br/>
<b>High Risk Assets:</b> {high_risk}<br/>
<b>Low Risk Assets:</b> {low_risk}<br/>
<b>Average Risk Score:</b> {avg_score}<br/>
<b>Harvest Now, Decrypt Later Risks:</b> {hndl}
"""

elements.append(
    Paragraph(summary, styles["BodyText"])
)

elements.append(Spacer(1,20))

# -----------------------------
# Table
# -----------------------------
elements.append(
    Paragraph(
        "<b>Risk Assessment Results</b>",
        styles["Heading2"]
    )
)

table_data = [
    [
        "Asset",
        "Algorithm",
        "Risk",
        "Score",
        "Recommendation"
    ]
]

for _, row in df.iterrows():

    table_data.append([
        row["Asset"],
        row["Algorithm"],
        row["RiskLevel"],
        row["RiskScore"],
        row["Recommendation"]
    ])

table = Table(table_data)

table.setStyle(
    TableStyle([
        ("BACKGROUND",(0,0),(-1,0),colors.darkblue),
        ("TEXTCOLOR",(0,0),(-1,0),colors.white),
        ("GRID",(0,0),(-1,-1),1,colors.black),
        ("BACKGROUND",(0,1),(-1,-1),colors.beige),
        ("ALIGN",(0,0),(-1,-1),"CENTER"),
        ("FONTNAME",(0,0),(-1,0),"Helvetica-Bold"),
        ("BOTTOMPADDING",(0,0),(-1,0),8),
    ])
)

elements.append(table)

elements.append(Spacer(1,20))

# -----------------------------
# Recommendations
# -----------------------------
elements.append(
    Paragraph(
        "<b>Recommendations</b>",
        styles["Heading2"]
    )
)

recommendations = """
• Replace RSA with ML-KEM Hybrid Deployment.<br/><br/>

• Replace ECC Digital Signatures with ML-DSA.<br/><br/>

• Continue using AES-256 for symmetric encryption.<br/><br/>

• Deploy Hybrid TLS during migration.<br/><br/>

• Prioritize long-term confidential systems for migration.<br/><br/>

• Improve key lifecycle management using HSMs and Key Vaults.
"""

elements.append(
    Paragraph(recommendations, styles["BodyText"])
)

elements.append(Spacer(1,20))

# -----------------------------
# Conclusion
# -----------------------------
elements.append(
    Paragraph(
        "<b>Conclusion</b>",
        styles["Heading2"]
    )
)

conclusion = """
The organization currently relies on several cryptographic systems
that are vulnerable to future quantum attacks.

A phased migration towards Post-Quantum Cryptography (PQC)
using hybrid cryptographic deployment is recommended to
maintain long-term confidentiality and regulatory compliance.
"""

elements.append(
    Paragraph(conclusion, styles["BodyText"])
)

# -----------------------------
# Build PDF
# -----------------------------
pdf.build(elements)

print("PDF Report Generated Successfully!")

print("Saved as:")

print("reports/QuantumSafe_Report.pdf")
```
