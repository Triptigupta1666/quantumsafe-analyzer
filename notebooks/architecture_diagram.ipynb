```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import os

# -----------------------------
# Create diagrams folder
# -----------------------------
os.makedirs("diagrams", exist_ok=True)

fig, ax = plt.subplots(figsize=(18,10))

ax.set_xlim(0,18)
ax.set_ylim(0,12)

ax.axis('off')

# -------------------------------------------------
# Helper Function
# -------------------------------------------------
def box(x,y,w,h,text,color="#D6EAF8"):

    rect = Rectangle(
        (x,y),
        w,
        h,
        linewidth=2,
        edgecolor="black",
        facecolor=color
    )

    ax.add_patch(rect)

    ax.text(
        x+w/2,
        y+h/2,
        text,
        ha='center',
        va='center',
        fontsize=10,
        weight='bold'
    )

# -------------------------------------------------
# Title
# -------------------------------------------------

plt.title(
    "Enterprise Hybrid Post-Quantum Cryptographic Architecture",
    fontsize=18,
    weight="bold"
)

# -------------------------------------------------
# Client Layer
# -------------------------------------------------

box(0.5,8.5,2.5,1.2,"Client\n(Web / Mobile)","#AED6F1")

box(4,8.5,3,1.2,"Hybrid TLS\nRSA + ML-KEM","#F9E79F")

box(8,8.5,2.5,1.2,"API Gateway","#AED6F1")

# -------------------------------------------------
# Application Layer
# -------------------------------------------------

box(12,9.3,4,1.2,"Internet Banking","#ABEBC6")

box(12,7.5,4,1.2,"Mobile Banking","#ABEBC6")

# -------------------------------------------------
# Security Layer
# -------------------------------------------------

box(6,5,3.5,1.3,"Identity Management\nML-DSA","#F5CBA7")

box(11.5,5,3.5,1.3,"Hardware Security Module\n(HSM)","#F5B7B1")

# -------------------------------------------------
# Data Layer
# -------------------------------------------------

box(2,1.5,4,1.4,"Customer Database\nAES-256","#D2B4DE")

box(8,1.5,4,1.4,"Cloud Storage\nAES-256","#D2B4DE")

box(14,1.5,3,1.4,"Backup Server","#D2B4DE")

# -------------------------------------------------
# Monitoring Layer
# -------------------------------------------------

box(6,10.8,6,0.8,"Security Operations Center (SOC)","#A9DFBF")

# -------------------------------------------------
# Draw Arrows
# -------------------------------------------------

arrow = dict(
    arrowstyle="->",
    lw=2
)

# Client -> TLS
ax.annotate("",xy=(4,9.1),xytext=(3,9.1),arrowprops=arrow)

# TLS -> API
ax.annotate("",xy=(8,9.1),xytext=(7,9.1),arrowprops=arrow)

# API -> Internet Banking
ax.annotate("",xy=(12,9.9),xytext=(10.5,9.2),arrowprops=arrow)

# API -> Mobile Banking
ax.annotate("",xy=(12,8.1),xytext=(10.5,9.0),arrowprops=arrow)

# Internet Banking -> Identity
ax.annotate("",xy=(8,6.3),xytext=(13.5,9.3),arrowprops=arrow)

# Identity -> HSM
ax.annotate("",xy=(11.5,5.65),xytext=(9.5,5.65),arrowprops=arrow)

# HSM -> Customer DB
ax.annotate("",xy=(5,2.9),xytext=(12.5,5),arrowprops=arrow)

# HSM -> Cloud
ax.annotate("",xy=(10,2.9),xytext=(13.2,5),arrowprops=arrow)

# HSM -> Backup
ax.annotate("",xy=(15.5,2.9),xytext=(14.5,5),arrowprops=arrow)

# SOC -> Banking
ax.annotate("",xy=(14,10.5),xytext=(11,10.8),arrowprops=arrow)

# -------------------------------------------------
# Layer Labels
# -------------------------------------------------

ax.text(-0.3,9.1,"Client Layer",fontsize=11,weight='bold')
ax.text(-0.3,5.7,"Security Layer",fontsize=11,weight='bold')
ax.text(-0.3,2.2,"Data Layer",fontsize=11,weight='bold')

# -------------------------------------------------
# Footer
# -------------------------------------------------

plt.figtext(
    0.5,
    0.01,
    "Hybrid Deployment: Classical Cryptography + Post-Quantum Cryptography",
    ha="center",
    fontsize=12,
    weight="bold"
)

# -------------------------------------------------
# Save
# -------------------------------------------------

plt.savefig(
    "diagrams/pqc_architecture_enterprise.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()

print("="*60)
print("Enterprise Architecture Diagram Generated Successfully")
print("="*60)

print("\nSaved As:")

print("diagrams/pqc_architecture_enterprise.png")
```
