```python
import os

# -----------------------------
# Create Reports Folder
# -----------------------------
os.makedirs("reports", exist_ok=True)

# -----------------------------
# Load Company Assets
# -----------------------------
df = pd.read_csv("assets.csv")

print("=" * 60)
print("QuantumSafe Analyzer")
print("Cryptographic Risk Assessment")
print("=" * 60)


# -----------------------------
# Function: Risk Assessment
# -----------------------------
def assess_algorithm(algorithm):

    algorithm = algorithm.upper()

    if algorithm == "RSA":
        return "High", 10, "Replace with ML-KEM Hybrid"

    elif algorithm == "ECC":
        return "High", 9, "Replace with ML-KEM + ML-DSA"

    elif algorithm == "AES":
        return "Low", 2, "Continue using AES-256"

    else:
        return "Unknown", 5, "Manual Review Required"


# -----------------------------
# Function:
# Harvest Now Decrypt Later
# -----------------------------
def harvest_now_risk(years, algorithm):

    if algorithm in ["RSA", "ECC"] and years >= 10:
        return "YES"

    return "NO"


# -----------------------------
# Lists
# -----------------------------
risk_level = []
risk_score = []
recommendation = []
harvest_risk = []

# -----------------------------
# Analyze Every Asset
# -----------------------------
for index, row in df.iterrows():

    risk, score, rec = assess_algorithm(row["Algorithm"])

    harvest = harvest_now_risk(
        row["DataLifeYears"],
        row["Algorithm"]
    )

    risk_level.append(risk)
    risk_score.append(score)
    recommendation.append(rec)
    harvest_risk.append(harvest)

    print("\n------------------------------------------")

    print("Asset:", row["Asset"])
    print("Algorithm:", row["Algorithm"])
    print("Risk:", risk)
    print("Risk Score:", score)
    print("Harvest Now Risk:", harvest)
    print("Recommendation:", rec)

# -----------------------------
# Add New Columns
# -----------------------------
df["RiskLevel"] = risk_level
df["RiskScore"] = risk_score
df["Recommendation"] = recommendation
df["HarvestNowDecryptLater"] = harvest_risk

# -----------------------------
# Save Report
# -----------------------------
df.to_csv("reports/risk_report.csv", index=False)

print("\n")
print("=" * 60)
print("Risk Assessment Completed Successfully")
print("Report Saved:")
print("reports/risk_report.csv")
print("=" * 60)
```
