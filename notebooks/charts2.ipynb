# Task 1 (continued) – Risk Visualization

Generates four charts from `reports/risk_report.csv` and saves them to `charts/` as PNGs, in addition to displaying them inline.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

os.makedirs("charts", exist_ok=True)

try:
    df = pd.read_csv("reports/risk_report.csv")
except FileNotFoundError:
    raise FileNotFoundError(
        "reports/risk_report.csv not found. Run risk_engine.ipynb first."
    )

print("Generating charts...")
```

```python
# -----------------------------
# Chart 1: Algorithm Distribution
# -----------------------------
algorithm_counts = df["Algorithm"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    algorithm_counts,
    labels=algorithm_counts.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Algorithm Distribution")
plt.savefig("charts/algorithm_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# -----------------------------
# Chart 2: Risk Level Distribution
# -----------------------------
risk_counts = df["RiskLevel"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(risk_counts.index, risk_counts.values)
plt.title("Risk Level Distribution")
plt.xlabel("Risk Level")
plt.ylabel("Number of Assets")
plt.savefig("charts/risk_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# -----------------------------
# Chart 3: Average Risk Score by Department
# -----------------------------
dept_risk = df.groupby("Department")["RiskScore"].mean()

plt.figure(figsize=(8, 5))
plt.bar(dept_risk.index, dept_risk.values)
plt.title("Average Risk Score by Department")
plt.xlabel("Department")
plt.ylabel("Average Risk Score")
plt.xticks(rotation=45)
plt.savefig("charts/department_risk.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# -----------------------------
# Chart 4: High Risk Assets
# -----------------------------
top_assets = df.sort_values(by="RiskScore", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(top_assets["Asset"], top_assets["RiskScore"])
plt.title("High Risk Assets")
plt.xlabel("Risk Score")
plt.savefig("charts/high_risk_assets.png", dpi=300, bbox_inches="tight")
plt.show()

print("\nAll charts saved to charts/")
```
